```python
import cv2
import pytesseract
import pathlib
import os
import platform
from matplotlib import pyplot as plt
import numpy as np
```

```python
address = "../examples/FFIFx3YXIAEL7xd.jpeg"
img = cv2.imread(address)
```

```python
def show_image(target):
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(target, cv2.COLOR_BGR2RGB), interpolation='nearest')
    plt.show()
```

```python
show_image(img)
```

```python
kernel = np.array([[-1,-1,-1], 
                    [-1, 9,-1],
                    [-1,-1,-1]])
sh = cv2.filter2D(img, -1, kernel)
show_image(sh)
```

```python
dst = cv2.fastNlMeansDenoisingColored(src=img, h=5)
show_image(dst)
```

```python
dst2 = cv2.fastNlMeansDenoising(src=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
show_image(dst2)
(a, flt) = cv2.threshold(dst2, 120, 255, type=cv2.THRESH_OTSU)
show_image(cv2.cvtColor(flt, cv2.COLOR_GRAY2BGR))
iv2 = np.invert(flt)
show_image(iv2)
pytesseract.image_to_string(dst2)
pytesseract.image_to_string(iv2)
```

```python
hsv = cv2.cvtColor(dst, cv2.COLOR_BGR2HSV)
show_image(hsv)
```

```python
mask = cv2.inRange(src=hsv, lowerb=(60, 40*255/100, 40*255/100), upperb=(80, 255, 65*255/100))
show_image(mask)
```

```python
bitwise = cv2.bitwise_and(dst, dst, mask=mask)
inv = np.invert(bitwise)
show_image(inv)
```

```python
m2 = cv2.adaptiveThreshold(cv2.cvtColor(dst,cv2.COLOR_BGR2GRAY),255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,3,15)
show_image(m2)
m3 = np.invert(m2)
show_image(m3)
```
